```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# setting one side of the grid

points=np.arange(-5,5,0.01) #step size 0.01
```

```python
#meshgrid
#help(np.meshgrid)
dx,dy = np.meshgrid(points,points)
dx #will be rows in the range of points 
dy #will be the columns in the range of points
```

```python
z = (np.sin(dx)+ np.sin(dy))
```

```python
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()

plt.title('sin(x)+sin(y)')
```

```python
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
condition = np.array([True,True,False,False])
```

```python
#method 1: List comprension to choose A if condition value is true. 

answer1 = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
```

```python
answer1
```

```python
# method 2: using numpy.where

answer2=np.where(condition,A,B) #if condition is True, choose A; otherwise choose B
```

```python
answer2
```

### Random Distn

```python
from numpy.random import randn
arr=randn(5,5) #randomly normalised 5 by 5 array

arr
```

```python
# using np.where to apply when a condition is met
#np.where is used to clean your data
np.where(arr<0,0,arr)# if the value of arr is <0 then change it to zero, else leave as is 
```

## Array Input&Output

```python
arr =np.arange(5)
```

```python
#set new arr
arr=np.arange(10)
```

```python
arr2=np.load('myfirstarr.npy')
```

```python
#to save multiple arrays
np.savez('ziparr.npz',x=arr,y=arr2)
```
